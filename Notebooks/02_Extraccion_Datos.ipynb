## Listar partidos liga colombiana

```python
import http.client
import json

def listar_partidos(id_league, temporada):
    conn = http.client.HTTPSConnection("www.thesportsdb.com")
    conn.request("GET", f"/api/v1/json/3/eventsseason.php?id={id_league}&s={temporada}")
    res = conn.getresponse()
    data = json.loads(res.read().decode("utf-8"))
    print(json.dumps(data, indent=2)) 
    return data

# Ejemplo de uso
partidos = listar_partidos("4497", "2025")
```
